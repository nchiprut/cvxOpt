```python
from cvxpy import *
import numpy as np

m, n = (30, 20)
np.random.seed(1)

A = np.asmatrix(np.random.randn(m, n))
b = np.asmatrix(np.random.randn(m, 1))
```

```python
x = Variable(n)

cost = sum_squares(A * x - b)

objective = Minimize(cost)
constraints = [0 <= x, x <= 1]
prob = Problem(objective, constraints)

result = prob.solve()
```

```python
print('Optimal x:\n', x.value)
print('Optimal lagrange multiplier:\n', 
      constraints[0].dual_value, 
      constraints[1].dual_value)
```

```python
# The optimal value for x is stored in x.value.
print(x.value)
# The optimal Lagrange multiplier for a constraint
# is stored in constraint.dual_value.
print(constraints[0].dual_value)
print(constraints[1].dual_value)
```

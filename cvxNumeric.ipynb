```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative

def non_convex(x):
    return x**3 + x**4

x = np.linspace(-1.5,1,100)
y = non_convex(x)
plt.plot(x, y, label="$x^3+x^4$")
plt.show()
```

```python
i = 0
theta = np.linspace(0,1,100)
is_violated = False
while not np.any(is_violated):
    i += 1
    rand_tuple = np.random.randint(0, x.size, 2)
    x_0 = x[rand_tuple]
    y_0 = y[rand_tuple]
    
    is_violated = theta*y_0[0] + (1 - theta)*y_0[1] < \
                    non_convex(theta*x_0[0] + (1 - theta)*x_0[1])
print('number of iterations: ', i)
```

```python
# Use labels to name lines in the graph.
plt.plot(x, y, label="$x^3+x^4$")
plt.plot(x_0, y_0, label=r"$\theta f(x)+(1-\theta )f(y)$")

# plt.xlabel and plt.ylabel assign names to the axes.
plt.xlabel(r"$x$", fontsize=16)
# plt.legend adds a legend to the graph with the line labels.
plt.legend(loc='upper left')

plt.show()
```

```python
i = 0
is_violated = False
while not np.any(is_violated):
    i += 1
    rand = np.random.randint(0, x.size)
    x_0 = x[rand]
    y_0 = y[rand]
    dx_0 = derivative(non_convex, x_0, 1e-5)
    
    tan = (x-x_0)*dx_0 + y_0
    is_violated = tan > y
print('number of iterations: ', i)
```

```python
# Use labels to name lines in the graph.
plt.plot(x, y, label="$x^3+x^4$")
plt.plot(x, tan, label=r"$f(x)+f'(x)(y-x)$")

# plt.xlabel and plt.ylabel assign names to the axes.
plt.xlabel(r"$x$", fontsize=16)
# plt.legend adds a legend to the graph with the line labels.
plt.legend(loc='upper left')

plt.show()
```
